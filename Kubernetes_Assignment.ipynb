📘 Introduction to Kubernetes — Assignment
1. Introduction

Kubernetes is an open-source container orchestration platform used to automate the deployment, scaling, and management of containerized applications. It helps developers manage applications efficiently across clusters of machines by providing features such as automated rollouts, self-healing, load balancing, and scaling.

In this assignment, a local Kubernetes cluster was created using Minikube, and a sample application was deployed. Resource management operations were performed, and Helm was used to package and deploy applications.

2. Kubernetes Setup

A local Kubernetes environment was set up using Minikube with Docker as the container runtime.

Steps Performed

Installed Docker Desktop.

Installed kubectl command-line tool.

Installed Minikube.

Started the Kubernetes cluster using:

minikube start --driver=docker

Verification

Cluster status was verified using:

kubectl get nodes


The node showed Ready status, confirming successful cluster creation.

3. Application Deployment

A sample application was deployed in Kubernetes using a container image.

Deployment Command
kubectl create deployment hello-minikube --image=kicbase/echo-server:1.0


This created a deployment with a running pod.

Exposing the Application

The deployment was exposed as a service using NodePort:

kubectl expose deployment hello-minikube --type=NodePort --port=8080


The application was accessed in the browser using:

minikube service hello-minikube


The browser displayed the response from the deployed application, confirming successful deployment.

4. Resource Management

Kubernetes resources such as Pods, Deployments, and Services were managed using kubectl commands.

Viewing Pods
kubectl get pods

Viewing Deployments
kubectl get deployments

Scaling the Deployment

The deployment was scaled from 1 replica to 3 replicas:

kubectl scale deployment hello-minikube --replicas=3


After scaling, multiple pods were observed running simultaneously.

This demonstrates Kubernetes auto-scaling capability.

5. Helm Charts

Helm is a package manager for Kubernetes that simplifies application deployment using predefined templates called charts.

Helm was installed and configured.

Adding Helm Repository
helm repo add bitnami https://charts.bitnami.com/bitnami
helm repo update

Installing NGINX Using Helm
helm install my-nginx bitnami/nginx

Verification
helm list
kubectl get pods


The NGINX application was successfully deployed and accessed in the browser using:

minikube service my-nginx


The browser displayed the NGINX welcome page, confirming successful Helm deployment.

6. Results

All components worked correctly and demonstrated Kubernetes functionality.
